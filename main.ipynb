Merging the dataset since the client dataset and price dataset likely contain information that is related to each other. Merging the datasets allows us to bring together all relevant information in one place for comprehensive analysis.

```python
import pandas as pd

# Load the datasets
client_data = pd.read_csv('client_data.csv')
price_data = pd.read_csv('price_data.csv')

# Merge the datasets on the 'id' column
merged_data = pd.merge(client_data, price_data, on='id', how='inner')

# Explore the merged data
merged_data.head()
```

```python
#Types of channels
channel_sales_counts = merged_data['channel_sales'].value_counts()
print(channel_sales_counts)
```

Distribution of electricity consumption over the past 12 months

```python
plt.hist(merged_data['cons_12m'], bins=20, range=(merged_data['cons_12m'].quantile(0.01), merged_data['cons_12m'].quantile(0.99)))
plt.xlabel('Electricity Consumption (past 12 months)')
plt.ylabel('Frequency')
plt.title('Distribution of Electricity Consumption (over past 12 months)')
plt.show()
```

 forecasted electricity consumption for the next 12 months and the next calendar year.

```python
plt.scatter(merged_data['forecast_cons_12m'], merged_data['forecast_cons_year'])
plt.xlabel('Forecasted Consumption (next 12 months)')
plt.ylabel('Forecasted Consumption (next calendar year)')
plt.title('Relationship between Forecasted Consumptions')
plt.xlim(merged_data['forecast_cons_12m'].quantile(0.01), merged_data['forecast_cons_12m'].quantile(0.99))
plt.ylim(merged_data['forecast_cons_year'].quantile(0.01), merged_data['forecast_cons_year'].quantile(0.99))
plt.show()
```

correlation between features

```python
merged_data['date_activ'] = pd.to_datetime(merged_data['date_activ'])
```

```python
correlation_matrix = merged_data.corr()
```

```python
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", annot_kws={"size": 10})
plt.title('Correlation Heatmap', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

net margin values for power subscriptions.

```python
plt.figure(figsize=(8, 6))
sns.histplot(merged_data['margin_net_pow_ele'], kde=True)
plt.xlabel('Net Margin on Power Subscription')
plt.ylabel('Count')
plt.title('Distribution of Net Margin')
plt.show()
```

Distribution of various features

```python
numeric_columns = ['cons_12m', 'cons_gas_12m', 'imp_cons', 'net_margin', 'pow_max']

sns.pairplot(merged_data[numeric_columns])
plt.suptitle('Pairwise Scatter Plot of Numerical Variables')
plt.show()
```

Total electrcity consumption for each month

```python
monthly_consumption = merged_data.resample('M', on='date_activ')['cons_12m'].sum()
plt.figure(figsize=(12, 6))
monthly_consumption.plot(kind='area', stacked=True)
plt.xlabel('Date')
plt.ylabel('Monthly Consumption')
plt.title('Monthly Electricity Consumption')
plt.show()
```

distribution of gas usage among customers

```python
has_gas_counts = merged_data['has_gas'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(has_gas_counts, labels=has_gas_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Gas Usage')
plt.axis('equal')
plt.show()
```

churn count based on different channel sales categories.

```python
cross_tab = pd.crosstab(merged_data['channel_sales'], merged_data['churn'])
cross_tab.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.xlabel('Channel Sales')
plt.ylabel('Count')
plt.title('Churn Count by Channel Sales')
plt.xticks(rotation=45)
plt.legend(title='Churn', loc='upper right')
plt.show()
```

Distribution of the forecasted discount on energy based on the churn status of customers

```python
plt.figure(figsize=(8, 6))
sns.lineplot(data=merged_data, x='churn', y='forecast_discount_energy')
plt.xlabel('Churn')
plt.ylabel('Forecasted Discount')
plt.title('Distribution of Forecasted Discount by Churn')
plt.show()
```

Finding how many customer has churned

```python
churn_counts = merged_data['churn'].value_counts()
total_customers = churn_counts.sum()
churn_percentage = churn_counts / total_customers * 100
plt.figure(figsize=(6, 6))
plt.bar(churn_counts.index, churn_counts.values)
plt.xlabel('Churn')
plt.ylabel('Count')
plt.title('Churned vs. Non-Churned Customers')
plt.xticks(churn_counts.index, ['Not Churned', 'Churned'])
plt.text(0, churn_counts[0] + 100, f'{churn_counts[0]} ({churn_percentage[0]:.1f}%)', ha='center')
plt.text(1, churn_counts[1] + 100, f'{churn_counts[1]} ({churn_percentage[1]:.1f}%)', ha='center')
plt.show()
```
